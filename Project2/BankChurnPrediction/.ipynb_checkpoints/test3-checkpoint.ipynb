```python
# === 1. NECESSARY IMPORTS ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV, learning_curve
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score, 
                             roc_curve, auc, precision_score, recall_score, f1_score, roc_auc_score)
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
```

```python
# === 2. LOADING THE DATASET ===
df = pd.read_csv('playground-series-s4e1/train.csv')
```

```python
# === 3. INITIAL EXPLORATORY ANALYSIS ===

# Display general information about the dataset:
# - Number of entries
# - Column names and data types
# - Non-null counts per column
print(df.info())

# Check for missing values in each column
# Important to determine whether data imputation or removal is needed
print(df.isnull().sum())

# Generate descriptive statistics for each numeric column:
# - Count, mean, standard deviation
# - Min, max, and quartiles (25%, 50%, 75%)
# Helps detect outliers and understand feature scales
print(df.describe())
```

```python
# === 4. DISTRIBUTION OF TARGET VARIABLE ===

# Plot a count plot of the target variable 'Exited'
# This shows how balanced the dataset is in terms of customers who churned (1) vs. those who didn't (0)
sns.countplot(x='Exited', data=df)

# Add a title to the plot for clarity
plt.title('Class Distribution (Churn vs Non-Churn)')

# Display the plot
plt.show()
```

```python
# === 5. PREPROCESSING ===

# Encode categorical variables: Gender and Geography
# Note: LabelEncoder assigns integers to categories, which can introduce order.
# It's acceptable for binary categories (e.g., Gender), but not ideal for nominal multi-class ones like Geography.
labelencoder = LabelEncoder()
df['Gender'] = labelencoder.fit_transform(df['Gender']) # Converts 'Male'/'Female' to 0/1
df['Geography'] = labelencoder.fit_transform(df['Geography']) # Converts 'France', 'Spain', 'Germany', etc. to 0, 1, 2 (may imply false ordering)

# Separate features (X) from target (y)
X = df.drop('Exited', axis=1) # 'Exited' is the target variable
y = df['Exited'] # Target variable: whether the customer churned (1) or not (0)

# Keep only numerical columns (encoded categorical + original numeric)
X = X.select_dtypes(include=[np.number])

# Normalize the features using StandardScaler (mean=0, std=1)
# This improves performance for models sensitive to feature scale (e.g., SVM, k-NN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# === 6. FEATURE SELECTION ===

# Use SelectKBest to select the top 10 features based on univariate statistical tests
# f_classif applies ANOVA F-test to determine feature importance for classification tasks
selector = SelectKBest(score_func=f_classif, k=10)

# Fit the selector and transform the scaled feature set
X_selected = selector.fit_transform(X_scaled, y)

# Get a boolean mask of selected features
selected_features_mask = selector.get_support()

# Retrieve the names of the selected features
selected_feature_names = X.columns[selected_features_mask]

# === Visualization of Feature Importance ===

# Create a barplot to visualize the importance (F-scores) of the selected features
plt.figure(figsize=(10, 6))

# Sort features by their F-score in descending order
sorted_idx = np.argsort(selector.scores_[selected_features_mask])[::-1]

# Plot scores with feature names
sns.barplot(x=selector.scores_[selected_features_mask][sorted_idx], 
            y=selected_feature_names[sorted_idx], palette="viridis")

# Add titles and labels for clarity
plt.title("Importance of Selected Features (SelectKBest)")
plt.xlabel("Score F (ANOVA)")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()

# Outcome: Identifies which variables are most predictive based on their statistical relationship with the target.
```

```python
# === 7. SPLITTING AND BALANCING ===

# Split the dataset into training and testing sets
# 70% for training, 30% for testing, with fixed random seed for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)

# === Handling Class Imbalance ===

# Initialize SMOTE (Synthetic Minority Over-sampling Technique)
# It synthetically generates new samples for the minority class
smote = SMOTE(sampling_strategy=0.5, random_state=42)

# Initialize RandomUnderSampler to reduce the majority class
under = RandomUnderSampler(sampling_strategy=0.8, random_state=42)

# Combine both oversampling and undersampling in a pipeline
pipeline = Pipeline([('o', smote), ('u', under)])

# Apply the pipeline to the training data to get a balanced dataset
X_train_balanced, y_train_balanced = pipeline.fit_resample(X_train, y_train)

# === Visualize class distribution after balancing ===

# Plot the new class distribution after SMOTE + undersampling
sns.countplot(x=y_train_balanced)
plt.title('Class Distribution after SMOTE + Undersampling')
plt.xlabel('Class (Exited)')
plt.ylabel('Number of Samples')
plt.show()

# Outcome: The dataset is now more balanced, improving model learning and reducing bias toward the majority class.
```

```python
# === 8. MODEL TRAINING AND EVALUATION ===

# === Decision Tree ===

# Define a grid of hyperparameters to search for the Decision Tree model
param_grid_dt = {'max_depth': [3, 5, 10, None], 'min_samples_split': [2, 5, 10], 'criterion': ['gini', 'entropy']}

# Use GridSearchCV to find the best combination of hyperparameters (5-fold cross-validation)
grid_dt = GridSearchCV(DecisionTreeClassifier(), param_grid_dt, cv=5, scoring='accuracy', n_jobs=-1)

# Measure training time for Decision Tree
start_train_dt = time.perf_counter()
grid_dt.fit(X_train_balanced, y_train_balanced)
end_train_dt = time.perf_counter()
train_time_dt = end_train_dt - start_train_dt

# Measure prediction time for Decision Tree
start_test_dt = time.perf_counter()
y_pred_dt = grid_dt.best_estimator_.predict(X_test)
end_test_dt = time.perf_counter()
test_time_dt = end_test_dt - start_test_dt

# === k-Nearest Neighbors ===

# Define grid for k-NN model
param_grid_knn = {'n_neighbors': [11,15,21], 'weights': ['uniform', 'distance'], 'metric': ['euclidean']}

# GridSearchCV to find best k-NN configuration
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, scoring='accuracy', n_jobs=-1)

# Measure training time for k-NN
start_train_knn = time.perf_counter()
grid_knn.fit(X_train_balanced, y_train_balanced)
end_train_knn = time.perf_counter()
train_time_knn = end_train_knn - start_train_knn

# Measure prediction time for k-NN
start_test_knn = time.perf_counter()
y_pred_knn = grid_knn.best_estimator_.predict(X_test)
end_test_knn = time.perf_counter()
test_time_knn = end_test_knn - start_test_knn

# === Support Vector Machine ===

"""# Define search space for SVM using RandomizedSearchCV
param_dist_svm = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': ['scale', 'auto']}

# RandomizedSearchCV is used here for faster search with fewer iterations
random_svm = RandomizedSearchCV(SVC(probability=True), param_distributions=param_dist_svm,
                                n_iter=6, cv=3, scoring='accuracy', random_state=42, n_jobs=-1)

# Measure training time for SVM
start_train_svm = time.perf_counter()
random_svm.fit(X_train_balanced, y_train_balanced)
end_train_svm = time.perf_counter()
train_time_svm = end_train_svm - start_train_svm

# Measure prediction time for SVM
start_test_svm = time.perf_counter()
y_pred_svm = random_svm.best_estimator_.predict(X_test)
end_test_svm = time.perf_counter()
test_time_svm = end_test_svm - start_test_svm"""

# Define hyperparameter search space
param_dist_svm = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Create RandomizedSearchCV for SVM
random_svm = RandomizedSearchCV(
    SVC(probability=True),
    param_distributions=param_dist_svm,
    n_iter=6,
    cv=3,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

# Measure training time
start_train_svm = time.perf_counter()
random_svm.fit(X_train_balanced, y_train_balanced)
end_train_svm = time.perf_counter()
train_time_svm = end_train_svm - start_train_svm

# Measure prediction time
start_test_svm = time.perf_counter()
y_pred_svm = random_svm.best_estimator_.predict(X_test)
end_test_svm = time.perf_counter()
test_time_svm = end_test_svm - start_test_svm

# Predict probabilities for ROC curve
y_prob_svm = random_svm.best_estimator_.predict_proba(X_test)[:, 1]

```

```python
# === 9. CLASSIFICATION REPORTS AND CONFUSION MATRICES ===

# Define a function to print classification metrics and show the confusion matrix
def plot_confusion_and_report(y_true, y_pred, model_name, color):
    # Print classification report:
    # Includes Precision, Recall, F1-Score, and Support for each class
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_true, y_pred, target_names=["No Churn", "Churn"]))

    # Generate the confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Plot the confusion matrix using a heatmap
    sns.heatmap(cm, annot=True, fmt='d', cmap=color, xticklabels=["No Churn", "Churn"], yticklabels=["No Churn", "Churn"])

    # Add labels and title for clarity
    plt.title(f"Confusion Matrix - {model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# Call the function for each model to display their evaluation results
plot_confusion_and_report(y_test, y_pred_dt, "Decision Tree", "Blues")
plot_confusion_and_report(y_test, y_pred_knn, "k-NN", "Greens")
plot_confusion_and_report(y_test, y_pred_svm, "SVM", "Reds")
```

```python
# === 10. ROC CURVES ===

# Get predicted probabilities for the positive class (Churn = 1) for each model
# These are needed to calculate the ROC curve
y_prob_dt = grid_dt.best_estimator_.predict_proba(X_test)[:, 1]
y_prob_knn = grid_knn.best_estimator_.predict_proba(X_test)[:, 1]
y_prob_svm = random_svm.best_estimator_.predict_proba(X_test)[:, 1]

# Compute False Positive Rate (FPR) and True Positive Rate (TPR) for each model
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_prob_knn)
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_prob_svm)

# Plot the ROC curves
plt.figure(figsize=(8,6))
plt.plot(fpr_dt, tpr_dt, label='Decision Tree')
plt.plot(fpr_knn, tpr_knn, label='k-NN')
plt.plot(fpr_svm, tpr_svm, label='SVM')

# Add diagonal reference line (baseline classifier)
plt.plot([0,1],[0,1],'k--')

# Set axis labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')

# Show legend to differentiate models
plt.legend()

# Display the plot
plt.show()

# Outcome: The ROC curve shows how well each model distinguishes between classes
# A curve closer to the top-left indicates better performance.
```

```python
# === 11. METRIC COMPARISON ===

# Define the model names for the chart
models = ['Decision Tree', 'k-NN', 'SVM']

# Calculate key performance metrics for each model
# These help compare how each model performs on the test set
precisions = [precision_score(y_test, y_pred_dt), precision_score(y_test, y_pred_knn), precision_score(y_test, y_pred_svm)]
recalls = [recall_score(y_test, y_pred_dt), recall_score(y_test, y_pred_knn), recall_score(y_test, y_pred_svm)]
f1s = [f1_score(y_test, y_pred_dt), f1_score(y_test, y_pred_knn), f1_score(y_test, y_pred_svm)]
aucs = [roc_auc_score(y_test, y_prob_dt), roc_auc_score(y_test, y_prob_knn), roc_auc_score(y_test, y_prob_svm)]

# Set bar positions and width for the grouped bar chart
x = np.arange(len(models)) # positions for the groups
width = 0.2 # width of each bar

# Create a grouped bar chart to visualize the metrics
fig, ax = plt.subplots(figsize=(10,6))

# Plot each metric (you can uncomment others as needed)
ax.bar(x - 1.5*width, precisions, width, label='Precision')
#ax.bar(x - 0.5*width, recalls, width, label='Recall')
#ax.bar(x + 0.5*width, f1s, width, label='F1-Score')
ax.bar(x + 1.5*width, aucs, width, label='AUC')

# Label the x-axis with model names
ax.set_xticks(x)
ax.set_xticklabels(models)

# Set the vertical axis range
ax.set_ylim(0, 1)

# Add axis labels and title
ax.set_ylabel('Score')
ax.set_title('Model Performance Comparison')

# Add legend to distinguish the metrics
ax.legend()

# Improve layout spacing
plt.tight_layout()

# Display the plot
plt.show()

# Outcome: Helps identify the most balanced model,
# particularly in terms of precision and class discrimination (AUC).
```

```python
# === 12. LEARNING CURVES ===

# Define a function to plot the learning curve for a given model
# This shows how training and validation scores change as the training size increases
def plot_learning_curve(model, X, y, title):
    # Calculate training and validation scores using cross-validation
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, n_jobs=-1)

    # Plot the average training score across folds
    plt.plot(train_sizes, np.mean(train_scores, axis=1), label='Training')

    # Plot the average validation score across folds
    plt.plot(train_sizes, np.mean(test_scores, axis=1), label='Validation')
    
    # Add title and axis labels
    plt.title(title)
    plt.xlabel('Training Size')
    plt.ylabel('Score')

    # Add legend and grid for clarity
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot learning curves for each model
plot_learning_curve(grid_dt.best_estimator_, X_train, y_train, "Learning Curve - Decision Tree")
plot_learning_curve(grid_knn.best_estimator_, X_train, y_train, "Learning Curve - k-NN")
plot_learning_curve(random_svm.best_estimator_, X_train, y_train, "Learning Curve - SVM")

# Outcome: Helps diagnose overfitting (high training, low validation score)
# or underfitting (both scores low), as the training set size increases.
```

```python
# === 13. LEARNING CURVES (MANUAL FEATURES) ===

# === Manual Feature Selection ===
# Select a subset of 5 important features manually
selected_manual_features = ['Age', 'NumOfProducts', 'IsActiveMember', 'Gender', 'Balance']

X_manual = df[selected_manual_features]
y_manual = df['Exited']

# Normalize the features
scaler_manual = StandardScaler()
X_manual_scaled = scaler_manual.fit_transform(X_manual)

# Split into training and test sets
X_train_manual, X_test_manual, y_train_manual, y_test_manual = train_test_split(
    X_manual_scaled, y_manual, test_size=0.3, random_state=42
)

from imblearn.pipeline import Pipeline

# Rebalance the training data using SMOTE + RandomUnderSampler
over = SMOTE(sampling_strategy=0.5, random_state=42)
under = RandomUnderSampler(sampling_strategy=0.8, random_state=42)
pipeline = Pipeline([('o', over), ('u', under)])
X_train_manual_balanced, y_train_manual_balanced = pipeline.fit_resample(X_train_manual, y_train_manual)


# === Decision Tree with Manual Features ===
param_grid_dt = {'max_depth': [3, 5, 10], 'min_samples_split': [2, 5], 'criterion': ['gini', 'entropy']}
grid_dt_manual = GridSearchCV(DecisionTreeClassifier(), param_grid_dt, cv=5, scoring='accuracy', n_jobs=-1)
grid_dt_manual.fit(X_train_manual_balanced, y_train_manual_balanced)

# Evaluate on test set
y_pred_dt_manual = grid_dt_manual.best_estimator_.predict(X_test_manual)

print("\n Decision Tree (with 5 manual features):")
print(classification_report(y_test_manual, y_pred_dt_manual, target_names=["No Churn", "Churn"]))

# Learning curve
train_sizes_dt, train_scores_dt, test_scores_dt = learning_curve(
    grid_dt_manual.best_estimator_,
    X_train_manual_balanced,
    y_train_manual_balanced,
    cv=5,
    n_jobs=-1
)

plt.plot(train_sizes_dt, np.mean(train_scores_dt, axis=1), label='Training Score')
plt.plot(train_sizes_dt, np.mean(test_scores_dt, axis=1), label='Validation Score')
plt.title("Learning Curve - Decision Tree (5 Features Manuais)")
plt.xlabel("Training Set Size")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()

# === k-NN with Manual Features ===
param_grid_knn_manual = {
    'n_neighbors': [11, 15, 21],
    'weights': ['distance'],
    'metric': ['euclidean']
}
grid_knn_manual = GridSearchCV(KNeighborsClassifier(), param_grid_knn_manual, cv=5, scoring='accuracy', n_jobs=-1)
grid_knn_manual.fit(X_train_manual_balanced, y_train_manual_balanced)

y_pred_knn_manual = grid_knn_manual.best_estimator_.predict(X_test_manual)

print("\n k-NN (with 5 manual features):")
print(classification_report(y_test_manual, y_pred_knn_manual, target_names=["No Churn", "Churn"]))

from sklearn.model_selection import learning_curve

train_sizes_manual, train_scores_manual, test_scores_manual = learning_curve(
    grid_knn_manual.best_estimator_,
    X_train_manual_balanced,
    y_train_manual_balanced,
    cv=5,
    n_jobs=-1
)

plt.plot(train_sizes_manual, np.mean(train_scores_manual, axis=1), label='Training Score')
plt.plot(train_sizes_manual, np.mean(test_scores_manual, axis=1), label='Validation Score')
plt.title("Learning Curve - k-NN (5 Manual Features)")
plt.xlabel("Training Set Size")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()

# Outcome: With only 5 manual features, both models show signs of overfitting, but still achieve around 80% accuracy with reasonable validation performance.
```

```python
# === 14. LEARNING CURVES (MANUAL FEATURES) ===

from imblearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, StratifiedKFold, learning_curve
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Create a pipeline combining SMOTE, undersampling, and SVM
# This ensures class balancing is done inside each fold of cross-validation
svm_pipeline = Pipeline([
    ('smote', SMOTE(sampling_strategy=0.5, random_state=42)),
    ('under', RandomUnderSampler(sampling_strategy=0.8, random_state=42)),
    ('clf', SVC(probability=True))
])

# Define hyperparameter grid for SVM (must prefix with "clf__" to target the pipeline step)
param_grid_svm = {
    'clf__C': [0.1, 1, 10],
    'clf__kernel': ['rbf'],
    'clf__gamma': ['scale']
}

# Use stratified k-fold to preserve class distribution in each fold
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform grid search with cross-validation over the full pipeline
grid_svm_manual = GridSearchCV(
    estimator=svm_pipeline,
    param_grid=param_grid_svm,
    cv=cv_strategy,
    scoring='accuracy',
    n_jobs=-1
)

# Train the model (no manual resampling needed  it's in the pipeline)
grid_svm_manual.fit(X_train_manual, y_train_manual)

# Predict on the test set
y_pred_svm_manual = grid_svm_manual.best_estimator_.predict(X_test_manual)

# Print performance metrics
print("\n SVM (with 5 manual features and pipeline):")
print(classification_report(y_test_manual, y_pred_svm_manual, target_names=["No Churn", "Churn"]))

# === Learning Curve for SVM ===
train_sizes_svm, train_scores_svm, test_scores_svm = learning_curve(
    grid_svm_manual.best_estimator_,
    X_train_manual,
    y_train_manual,
    cv=cv_strategy,
    n_jobs=-1
)

# Plot learning curve
plt.plot(train_sizes_svm, np.mean(train_scores_svm, axis=1), label='Training Score')
plt.plot(train_sizes_svm, np.mean(test_scores_svm, axis=1), label='Validation Score')
plt.title("Learning Curve - SVM (5 Manual Features - Pipeline)")
plt.xlabel("Training Set Size")
plt.ylabel("Score")
plt.legend()
plt.grid(True)
plt.show()

# Outcome: SVM generalizes well with balanced pipeline and 5 features, showing stable validation performance.
```
